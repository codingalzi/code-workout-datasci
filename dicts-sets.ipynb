비순차 자료형은 문자열, 리스트, 튜플과는 달리 순서를 따지지 않으며 항목의 중복도 허용하지 않는다.

## 사전 

사전은 **키**<font size = "2">key</font>와 **값**<font size = "2">value</font>의
쌍으로 구성된 항목들로 구성된 모음 자료형이며, **딕셔너리**<font size = "2">dictionary</font> 라고도 불린다.

예를 들어,  `'Hello':'안녕'`은 `'Hello'`를 키로, `'안녕'`을 값으로 갖는 쌍이며,
`'world':'세계'`는 `'world'`를 키로, `'세계'`를 값으로 갖는 쌍이다.
두 쌍을 항목으로 갖는 사전은 다음과 같이 두 쌍을 중괄호로 감싼다.

```python
{'Hello':'안녕', 'world':'세계'}
```

**빈 사전**

빈 사전은 아무것도 포함하지 않는 사전을 의미한다. 
빈 사전를 만드는 방법은 아래와 같으며, 
중괄호를 사용한다.

````python
empty_dict = {}
empty_dict
````

다음 방식도 빈 사전을 생성한다.

````python
empty_dict = dict()
empty_dict
````

**`dict` 자료형**

사전은 `dict` 자료형의 값이다.

````python
type({'Hello':'안녕', 'world':'세계'})
````

**`dict()` 함수**

`dict()` 함수는 키와 값의 튜플 형식의 항목으로 구성된 모음 자료형을
사전 자료형으로 변환한다. 

예를 들어, 아래 코든느 튜플로 구성된 리스트를 사전으로 변환한다.
튜플의 첫째 항목은 키로, 둘째 항목은 값으로 사용된다.

````python
data = [('Hello', '안녕'), ('World', '세계'), ('Programming', '프로그래밍')]
dict(data)
````

`zip()` 함수와 `dict()` 함수를 조합하여 편리하게 사전을 생성할 수 있다.

````python
data = zip('abcde', [1, 2, 3, 4, 5])
data_dict = dict(data)
data_dict
````

### 사전의 키 자료형

사전의 키로 정수, 부동소수점, 문자열, 튜플 등 불변 자료형만 허용된다.
즉, 리스트와 같은 가변 자료형은 사전의 키로 사용될 수 없다.
반면에 키에 대응하는 값은 임의의 자료형이 허용된다.

- 정수, 부동소수점 키

````python
dic1 = {2: '정수', 1.2 : '부동소수점'}
dic1
````

- 문자열 키

````python
dic2 ={'a':1, 'b':2, 'c':3}
dic2
````

- 튜플 키

````python
dic3 = {(1, 'a') : [1, 2, 3], (2, 'b') : [4, 5, 6]}
dic3
````

- 리스트 키?

리스트를 키로 사용할 수는 없다는 의미에서 `TypeError`가 발생한다.

````python
dic4 = {[1, 2] : 'a'}
dic4
````

- 리스트를 포함하는 튜플은?

리스트와 같은 가변 자료형은 어떤 형식으로도 키에 포함되어서는 안된다.
아래 코드는 튜플을 키로 사용하지만 키의 첫째 항목이 리스트이기에 허용되지 않는다.

````python
dic4 = {([1, 2], '가') : 'ABC'}
dic4
````

### 사전 연산

문자열, 리스트, 튜플과는 다르게 사전 연산은 특별히 없다.
다만 항목 포함 여부와 항목의 개수를 확인해주는 `in` 연산자와 
`len()` 함수 정도만 활용된다.

**`in` 연산자**

사전의 키로 사용되었는 여부를 알려주는 논리 연산자다.
예를 들어, 아래 코드는 `'city'` 문자열이 사전의 키로 사용되었는지 여부를 판별한다.

````python
'city' in {"name": "강현", "age": "3"} 
````

`in`의 부정은 `not in`을 사용한다. 즉
키로 사용되지 않았는지 여부를 뭍는다.

````python
'city' not in {"name": "강현", "age": "3"} 
````

**`len()` 함수**

사전에 포함된 항목의 개수를 반환한다.

````python
len({"name": "강현", "age": "3"}) 
````

### 사전 인덱싱 

사전 인덱싱은 키를 활용한다.
사전은 비순차 자료형이기에 정수 인덱스를 지원하지 않으며 따라서 정수 인덱싱도 사용할 수 없다.

예를 들어, 아래 딕셔너리 `dic`에서 `'Hello'` 문자열을 키로 갖는 항목의 값을 확인하고자 하면 
다음과 같이 대괄호 안에 `'Hello'`를 지정한다.

````python
dic = {'Hello' : '안녕', 'World' : '세계'}
dic['Hello']
````

만약에 키로 사용되지 않는 값으로 인덱싱을 시도하면 
지정된 키가 없다는 `KeyError` 오류가 발생한다. 

````python
dic['Python']
````

사전은 가변 자료형이다. 
즉, 사전의 항목을 추가, 업데이트, 삭제할 수 있다.

**사전 항목 추가, 업데이트**

사전에 항목을 추가하거나 기존 항목을 업데이트는 사전 인덱싱 형식으로 이루어진다.

```python
사전[key] = value
```   

예를 들어, `dic`에 `Python : 파이썬` 항목을 다음과 같이 추가할 수 있다.

````python
dic['Python'] = '파이썬'
dic
````

만약에 키가 이미 사용되었다면 키와 연결된 값이 업데이트 된다.
예를 들어 아래 코드는 `'world'` 키의 값의 `'세계'`에서 `'세상'`으로 업데이트한다.

````python
dic['World'] = '세상'
dic
````

**사전 항목 삭제**

사전의 항목은 `del` 명령어에 키를 지정하여 삭제할 수 있다.
아래 코드는 `'world'` 가 키로 사용된 항목을 삭제한다.

````python
del dic['World']
dic
````

사용되지 않은 키를 지정해서 항목 삭제를 시도하면 
해당 키를 갖는 항목이 없다는 `KeyError` 오류가 발생한다.

````python
del dic['World']
dic
````

    - `items()`

언급된 메서드의 기능을 설명을 위해 아래 사전을 이용한다.

````python
dic = {'Hello' : '안녕', 'World' : '세계'}
````

**`keys()` 메서드**

사전에 사용된 키들로 이루어진 `dict_keys`라는 모음 자료형을 반환한다.
`dict_keys` 자료형의 값은 리스트는 아니지만 반복문 등에서 유사하게 활용될 수 있다.

````python
dic.keys()
````

**`values()` 메서드**

사전에 사용된 값들으로 이루어진 `dict_values`라는 모음 자료형을 반환한다.
반환값은 역시 반복문 등에서 리스트처럼 유용하게 활용된다.

````python
dic.values()
````

**`items()` 메서드**

사전에 사용된 키와 값의 쌍으로 이루어진 `dict_items`라는 모음 자료형을 반환한다.
반환값은 (키, 값) 형식의 항목으로 구성된 리스트처럼 반복문 등에서 활용될 수 있다.

````python
dic.items()
````

**`get()` 메서드**

사전 인덱싱과 유사하게 작동한다.
즉, 메서드의 인자로 지정된 키에 해당하는 값을 반환한다.
아래 코드는 `dic['Hello']`와 동일한 값을 반환한다.

````python
dic.get('Hello')
````

사전 인덱싱과는 다르게 
메서드의 인자가 키로 사용되지 않았다 하더라도 오류가 발생하지는 않고
대신 `None`을 반환한다.

````python
print(dic.get('hello'))
````

`get()` 메서드의 둘째 인자를 지정하면 사전의 키로 사용되지 않은 인자에 대해
`None`이 아닌 둘째 인자를 대신 반환한다.

````python
dic.get('hello', "해당 키가 없어요.")
````

### 사전 반복문

사전에 대한 반복문은 키에 대해 실행된다.
아래 코드는 `data_dict` 사전에 사용된 키들만을 화면에 줄바꿈 없이 출력한다.

````python
data_dict = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5}

for item in data_dict:
    print(item, end=' ')
````

**`items()` 메서드 활용**

키와 값의 쌍에 대해 반복문을 실행하려면 `items()` 메서드를 이용한다. 
단, 키와 값을 따로따로 화면에 출력하는 아래 코드에서처럼 튜플 해체 기법으로 키와 값 각각에 대해 변수를 지정하는 게 편리하다.

````python
for key, value in data_dict.items():
    print(f"{key:>8} 키의 값: {value}")
````

물론 튜플 해체를 반드시 사용할 필요는 없다.
다만 필요에 따라 튜플 인덱싱으로 키와 값을 구분에서 활용하면 된다.

````python
for item in data_dict.items():
    key = item[0]
    value = item[1]
    print(f"{key:>8} 키의 값: {value}")
````

**`values()` 메서드 활용**

아래 코드처럼 값만을 이용하려면 `values()` 메서드와 함께
반복문을 싱행한다.

````python
for item in data_dict.values():
    print(item, end=' ')
````

그런데 리스트와 같은 가변 자료형은 키에 어떤 형식으로 포함되지 않아야 한다.

**예제 2**

정수들의 리스트가 다음과 같이 주어졌다.

````python
num_list = [2, 5, 2, 3, 3, 2]
````

(1) `num_list`에서 정수 2가 항목으로 등장하는 모든 인덱스로 구성된 리스트를 가리키는 `idx_2` 변수를 정의하는 코드를 작성하라.

힌트: `enumerate()` 함수와 `for` 반복문 활용

답:

`enumerate()` 함수를 리스트에 적용하면 항목과 인덱스의 쌍으로 구성된 모음 자료형이 생성되고,
이 점을 활용하여 `for` 반복문을 실행하면 항목과 인덱스의 관계가 확인된다.

````python
for idx, item in enumerate(num_list):
    print(f"{idx} 번째 요소: {item}")
````

유사한 방식으로 정수 2가 위치한 곳의 인덱스만 추출하여 새로운 리스트를 생성하면 된다.

````python
idx_list = []
for idx, item in enumerate(num_list):
    if item == 2:
        idx_list.append(idx)

print(idx_list)
````

(2) `num_list`에 포함된 항목을 인자로 받아, 그 항목이 등장하는 모든 인덱스를 리스트로 반환하는 `idx_list_func()` 함수를 정의하라.

답:

이전 문제의 코드를 함수의 본문으로 사용하기 위해 함수의 매개변수와 반환값만 지정하면 된다.

- 함수 매개변수: 정수 2를 매개변수화 해야 함
- 함수 반환값: `idx_list`를 반환값으로 지정

````python
def idx_list_func(target):
    idx_list = []
    for idx, item in enumerate(num_list):
        if item == target:
            idx_list.append(idx)
    return idx_list
````

정수 2가 위치한 인데스들의 리스트는 다음과 같다.

````python
idx_list_func(2)
````

정수 5가 위치한 인데스들의 리스트는 다음과 같다.

````python
idx_list_func(5)
````

(3) 이전 문제의 `idx_list_func()` 함수를 수정하여 두 개의 인자를 받도록 하라.

- 첫째 매개변수 `num_list`: 임의의 리스트
- 둘째 매개변수 `target`: `num_list`의 항목

함수 반환값은 `num_list`에서 `target`이 등장하는 모든 인덱스로 구성된 리스트이다.

답:

이전 문제의 `idx_list_func()` 함수에서 `num_list`는 고정된 리스트였는데 이를 매개 변수로 바꾸면
임의의 리스트에 대해 함수가 동일한 방식으로 작동한다.

````python
def idx_list_func(num_list, target):
    idx_list = []
    for idx, item in enumerate(num_list):
        if item == target:
            idx_list.append(idx)
    return idx_list
````

함수 호출을 위해 이제는 `num_list` 매개 변수에 대한 인자도 지정해야 한다.

````python
idx_list_func(num_list, 2)
````

다른 리스트를 인자로 지정해도 된다.

````python
num_list_2 = list('abcdeccea')
num_list_2
````

````python
idx_list_func(num_list_2, 'c')
````

```python
{'김강현': ['010-1234-5678', 20, 172.5, '제주'],

 '남세원': ['010-3456-7891', 21, 156.7, '경기'],

 '김현선': ['010-3333-8888', 22, 164.6, '광주'],
```

 힌트: 리스트 해체, `for` 반복문, 사전 항목 추가

6명 각자의 이름을 키로, 나머지 정보로 구성된 리스트를 값으로 사용하는 사전을 구성한다.

```python
name, *info = person
```

그러면 `name`은 이름을, `info`는 나머지 항목으로 구성된 리스트를 가리키게 되어

```python
name = person[0]
info = person[1:]
```

`info_dict[name] = info`

````python
info_dict = dict()

for person in info_list:
    name, *info = person
    info_dict[name] = info

info_dict
````

(3) `info_tel()` 함수를 `get()` 사전 메서드의 반환값을 활용하도록 함수를 재정의하라.
단, `info_dict` 사전의 키로 사용되지 않은 경우 `'정보 없음!'` 문자열을 반환해야 한다.

답:

이전 정의에서 `info_dict[name]` 표현식을
`info_dict.get(name)` 표현식으로 대체하면 일단 절대로 오류가 발생하지 않는다.
그리고 `if`조건문을 활용하여 `info_dict.get(name)`의 반환값이 `None`인 경우
`'정보 없음!'` 문자열을 반환하도록 하면 된다.

아래 코드에서 `if info`는 `info`가 가리키는 값이 `None`이 아닌 경우에 해당한다.

````python
def info_tel(name):
    info = info_dict.get(name)
    if info:
        return info[0]
    else:
        return "정보 없음!"
````

### 집합 활용법

프로그래밍에서 집합은 주로 순차 자료형의 항목에서 중복을 제거하고 싶을 때 사용한다.
이때 `set()` 함수가 매우 유용하다.

예를 들어 아래 코드는 리스트에서 중복된 항목이 삭제된, 
하지만 동일한 항목을 포함한 새로운 리스트를 생성한다.
단, 기존 리스트에서 사용된 항목들의 순서는 보존되지 않음에 주의한다.

````python
list(set([5, 1, 1, 2, 5, 5, 1]))
````

### 집합 연산

튜플 연산은 사전의 경우와 유사하게 작동한다.

**`in` 연산자**

집합의 항목(원소)으로 등장하는지 여부를 알려주는 논리 연산자다.

````python
1 in {1, 2, 3, 9, 4}
````

````python
'a' not in {1, 'b', True, 9}
````

**`len()` 함수**

집합에 포함된 항목의 개수를 반환한다.
단, 중복은 허용하지 않는다.

````python
len({1, 3, 5, 5, 7, 5, 7, 9})
````

단, 중복 항목은 하나로 간주한다.

## 조건제시법


조건제시법<font size = "2">comprehension</font>은 기존의 여러 값으로부터 일정한 규칙이나 조건을 만족하는 값들을 골라 새로운 모음 자료형을 생성하는 표현식이며,
컴프리헨션, 또는 내포라고 불리기도 한다.
조건제시법을 이용하여 리스트, 집합, 사전, 튜플 등을 정의할 수 있다.

### 집합 조건제시법

수학에서 0과 10사이에 있는 홀수들의 제곱을 원소로 갖는 집합을 조건제시법으로 표현하면
다음과 같다.

$$\{ x^2 \mid 0 \le x \le 10, \text{ 단 $x$는 홀수} \}$$

0과 10 사이에 있는 홀수들의 제곱을 항목으로 갖는 집합을 다음과 같이 정의할 수 있다.

````python
zero2ten_odd_set = {x**2 for x in range(11) if x%2 == 1}
zero2ten_odd_set
````

위 코드는 실제로는 다음 코드와 동일하게 파이썬 내부에서 실행된다.
아래 코드에서 `add()`는 집합에 항목을 추가하는 집합 메서드이다.

````python
zero2ten_odd_set = set()

for x in range(11):
    if x % 2 == 1:
        zero2ten_odd_set.add(x**2)

zero2ten_odd_set
````

**`if` 조건문의 역할**

`if` 조건문을 필요에 따라 사용하지 않아도 된다.
아래 코드는 0부터 10 사이의 모든 정수들의 제곱으로 구성된 집합을 생성한다.

````python
zero2ten_set = {x**2 for x in range(11)}
zero2ten_set
````

### 리스트 조건제시법

조건제시법으로 0부터 10 사이에 있는 홀수들의 제곱을 항목으로 갖는 리스트를 집합 조건제시법과
사실상 동일하게 생성할 수 있다.

````python
zero2ten_odd_list = [x**2 for x in range(11) if x%2 == 1]
zero2ten_odd_list
````

위 코드를 반복문으로 구현하면 다음과 같다.

````python
zero2ten_odd_list = []

for x in range(11):
    if x%2 == 1:
        zero2ten_odd_list.append(x**2)

zero2ten_odd_list
````

아래 코드는 `if` 조건문을 사용하지 않아서
최종적으로 0부터 10 사이의 모든 정수들의 제곱으로 구성된 리스트를 생성한다.

````python
zero2ten_list = [x**2 for x in range(11)]
zero2ten_list
````

### 사전 조건제시법

조건제시법을 이용하여 사전을 생성하는 과정도 유사하다. 
아래 코드는 0부터 10 사이의 홀수를 키로, 홀수의 제곱을 값으로 갖는 항목으로 구성된 사전을 생성한다.

````python
zero2ten_odd_dict = {x : x**2 for x in range(11) if x%2 == 1}
zero2ten_odd_dict
````

아래 코드는 `if` 조건문을 사용하지 않아서
최종적으로 0부터 10 사이의 모든 정수와 정수들의 제곱의 쌍으로 구성된 사전을 생성한다.

````python
zero2ten_dict = {x: x**2 for x in range(11)}
zero2ten_dict
````

### 튜플 조건제시법

조건제시법을 이용하여 튜플을 생성하는 과정은 `tuple()` 함수를 함께 사용한다는 점에서 조금 다르다.
엄밀히 말해 `tuple()` 함수의 인자로 사용되는 표현식은 튜플 조건제시법이 아니라 
[제너레이터 표현식](https://codingalzi.github.io/pybook/iterator_generator.html)이며,
여기서는 자세히 설명하지 않고, 편의상 튜플 조건제시법이라고 부른다.

예를 들어 아래 코드는 0부터 10 사이의 홀수들의 제곱으로 구성된 튜플을 생성한다.

````python
zero2ten_odd_tuple = tuple(x**2 for x in range(11) if x%2 == 1)
zero2ten_odd_tuple
````

아래 코드는 `if` 조건문을 사용하지 않아서
최종적으로 0부터 10 사이의 모든 정수들의 제곱으로 구성된 튜플을 생성한다.

````python
zero2ten_tuple = tuple(x**2 for x in range(11))
zero2ten_tuple
````

**예제 4: 튜플 생성**

아래 리스트는 섭씨 온도로 구성되어 있다.

````python
celsius = [0, 10, 20, 30]
````

`tuple()` 함수를 이용하여 화씨 온도로 구성된 튜플을 생성하라.
화씨 온도는 다음 공식으로 계산한다.

```python
fahrenheit = celsius * 9 / 5 + 32
```

실행 결과는 다음과 같아야 한다.

```python
(32.0, 50.0, 68.0, 86.0)
```

답:

`tuple()` 함수에 제너레이터 표현식을 전달하면, 생성된 값들을 튜플로 묶을 수 있다.
여기서는 섭씨 온도 `temp`를 하나씩 꺼내 화씨 온도 `temp * 9 / 5 + 32`로 변환한다.

`0`, `10`, `20`, `30`도는 각각 화씨 `32.0`, `50.0`, `68.0`, `86.0`도이다.
따라서 결과는 `(32.0, 50.0, 68.0, 86.0)`이다.

````python
fahrenheit_tuple = tuple(temp * 9 / 5 + 32 for temp in celsius)
fahrenheit_tuple
````

섭씨를 화씨로 변환하는 함수를 이용할 수도 있다.

````python
# 섭씨 => 화씨 변환 함수
def celsius_to_fahrenheit(celsius):
    return celsius * 9 / 5 + 32

fahrenheit_tuple = tuple(celsius_to_fahrenheit(temp) for temp in celsius)
fahrenheit_tuple
````
