**문제 5.** 다음 코드를 실행하면 `r ≈ 0.0000`이 출력된다. 이 결과의 해석으로 가장 적절한 것은?
단, `np.corrcoef(x, y)`는 두 배열 `x`, `y`의 피어슨 상관계수 행렬을 반환한다. 
결과 행렬의 `[0, 1]` 원소가 x와 y 사이의 상관계수이다.

```python
x = np.linspace(-1, 1, 200)    # -1부터 1까지 200개의 균등 간격 값
y = x ** 2                      # 이차함수 (포물선)
r = np.corrcoef(x, y)[0, 1]
print(f"상관계수: {r:.4f}")     # 약 0.0000
```

- (A) `x`와 `y`는 독립적이므로 분석에서 제외해야 한다.
- (B) 제곱 연산은 항상 r = 1을 가져야 하므로 계산 오류이다.
- (C) x의 범위가 −1 ~ 1로 너무 좁아 신뢰할 수 있는 상관계수 계산이 불가능하다.
- (D) 두 변수 간 선형 관계가 없으면 공분산도 반드시 0이다.
- (E) `r ≈ 0`이어도 포물선 형태의 강한 비선형 패턴이 존재하므로, 산점도를 직접 확인해야 한다.

**정답: (E)**

해설: `x`가 −1 ~ 1 사이에 **대칭**으로 분포하므로, x의 양수 구간에서는 y = x²가 x와 같은 방향으로, 음수 구간에서는 반대 방향으로 변화해 공분산이 상쇄되어 `r ≈ 0`이 된다. 그러나 산점도를 그리면 명확한 포물선(U자) 형태가 보인다. 피어슨 상관계수는 **선형 관계만** 측정하므로, `r ≈ 0`이 '아무 관계 없음'을 뜻하지는 않는다.

**문제 21.** 다음 코드의 **라인 A**에서 `(X - X.mean())`의 연산이 가능한 이유를 설명하시오.

```python
X = np.array([170.0, 175.0, 160.0, 180.0, 165.0])  # 키 (cm)
Y = np.array([ 65.0,  70.0,  55.0,  80.0,  60.0])  # 몸무게 (kg)

cov_xy = np.mean((X - X.mean()) * (Y - Y.mean()))   # A
r      = cov_xy / (X.std() * Y.std())               

print(f"공분산    : {cov_xy:.2f}")
print(f"상관계수  : {r:.4f}")
```

**정답:**

`X`는 shape `(5,)` 배열이고 `X.mean()`은 스칼라이다. 스칼라는 넘파이 브로드캐스팅 규칙에 의해 배열과 같은 shape으로 자동 확장되어 원소별 뺄셈이 가능하다.

**문제 22.** 다음 코드에서 `.agg(['mean', 'std'])` 대신 `.mean()`만 쓰면 `result`가 어떻게 달라지는지 설명하시오.

```python
df = pd.DataFrame({
    'Gender': ['Male', 'Male', 'Female', 'Female', 'Female'],
    'Score' : [  85,     92,      78,       88,       95  ]
})

result = df.groupby('Gender')['Score'].agg(['mean', 'std'])
print(result)
```
