#**7장. 이항분포**

##**이항분포**
동일한 시도를 n번 시도할 때, 특정 사건이 r번 나올 확률들의 분포를 **이항분포**라 한다.
> *조건* : 각각의 시도는 상호 독립적이며, 각각의 시도에서 특정 사건이 발생할 확률은 동일

✏️ **예제**) 4지선다 문제    
하나의 질문을 임의로 찍었을 때 맞출 확률 $= \frac{1}{4}$    
```
질문❔   
n개의 질문을 모두 임의로 찍어서 맞출 때 r개의 정답을 얻을 확률?
```

|$r$|0|1|…|$r$|…|n-1|n|
|:-:|:-:|:-:|:-|:-:|:-:|:-:|:-|
|$P(X=r)$||||**P(X=r)을 어떻게 계산?**||||
> $r$ : 맞히는 개수     
> $P(X=r)$ : r개 맞힐 확률

---

✏️ $P(X=r)$ **계산 예제** : <u>n=3인 경우</u>    
$\rightarrow$ 즉, 3개의 문제 중에 r개를 맞힐 확률    
먼저 $rϵ${0,1,2,3} $\rightarrow$ 4가지 경우를 살펴보아야 한다
1. **모두 틀린 경우**    
$P(X=0)=(\frac{3}{4})^3=^{3}C_{0}⋅(\frac{1}{4})^0⋅(\frac{3}{4})^3 $
2. **한 문제 맞힐 경우**   
$P(X=1)=^{3}C_{1}⋅(\frac{1}{4})^1⋅(\frac{3}{4})^2 $
3. **두 문제 맞힐 경우**   
$P(X=2)=^{3}C_{2}⋅(\frac{1}{4})^2⋅(\frac{3}{4})^1 $
4. **세 문제 맞힐 경우**    
$P(X=3)=^{3}C_{3}⋅(\frac{1}{4})^3⋅(\frac{3}{4})^0 $

$$\Rightarrow P(X=r)=^{3}C_{r}⋅(\frac{1}{4})^r⋅(\frac{3}{4})^{3-r} $$

----

##**일반화**
한 번의 시도에서 <u>성공</u>할 확률 = $P$
> *성공 : 원하는 결과가 발생한다는 의미*    

$$ n번\ 시도에서\ r번\ 성공할\ 확률   $$ 
$$ P(X=r)=^{n}C_{r}⋅p^r⋅(1-p)^{n-r} $$    
$$= ^{n}C_{r}⋅p^r⋅(q)^{n-r}$$    
$$(q=1-p)$$
> - $p^r⋅(1-p)^{n-r}$ 🟰 $^{n}C_{r}$ 경우 각각이 발생할 확률     
- $^{n}C_{r}$ 🟰 n개 중에서 r개를 선택하는 방법의 수    
- $p^r$ 🟰 선택한 r개가 성공할 확률     
- $q^{n-r}$ 🟰 r개를 제외한 나머지가 실패할 확률

이런 경우 X는 **'이항분포를 따른다'**고 말하고 아래와 같이 표기    
  $$X~\sim B(n,p)$$
> - $B$ 🟰 이항분포     
- $n$ 🟰 반복횟수    
- $p$ 🟰 한 번의 시도에서 성공할 확률 

---

✏️ **예제**) 이항분포 적용 연습 (*교재에 없음*)    
```
질문 ❔    
하나의 정상적인 동전을 30번 던져서 24번 앞면이 나올 확률은?
````

```
견본 답안 ❕    
정상적인 동전을 30번 던져서    
앞면이 나오는 횟수에 대한 확률분포 X는 이항분포를 따르며,     
아래 관계를 만족시킨다.
````
$$X~\sim B(30,\frac{1}{2})$$
```
따라서 앞면이 24번 나올 확률은 다음과 같다
```
$$ P(X=24)=^{30}C_{24}⋅(\frac{1}{2})^{24}⋅(\frac{1}{2})^{6} =^{30}C_{24}⋅\frac{1}{2}^{30} \approx 0.0006$$ 

```
또한 앞면이 24회 이상 나올 확률은 다음과 같다
```
$$ P(X\geq24)=P(X=24) + … + P(X=30) = 0.0007$$


**결론** (*이후에 **가설검정**을 배울 때 자세히 다룬다*)    
👉 일반적으로 사건 발생확률이 **5% 이하**면, <u>그런 사건이 실제로 발생하기 어렵다</u>고 본다.     
  따라서 동전을 30번 던져서 앞면이 24번 나온다면 그 동전은 정상적이지 않다고 판단한다.

----

##**이항분포의 기대치와 분산**
$$X~\sim B(n,p)라\ 하자$$     
- $E(X) = n ⋅ p$
- $Var(X) = n ⋅ p ⋅ q\ \ \ \ \  _{(q=1-p)}$


### **설명 1**    
$X_{1}$을 한 번 시도한 경우의 확률분포라 하자    

|$r$|0|1|
|:-:|:-:|:-:|
|$P(X=r)$|q|p|   
한 번 시도하면 **성공** 또는 **실패**만 발생    

**결론**    
$E(X_{1}) = 0 ⋅ q + 1 ⋅ p = p $    
$ Var(X_{1}) = (0-p)^2 ⋅ q + (1-p)^2 ⋅p$      
$\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ =p^2 ⋅(1-p)+(1-2p+p^2)⋅p$      
$\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ =p^2-p^3+p-2p^2+p^3$     
$\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ =p-p^2=p(1-p)$     
$\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ =p⋅q$


### **설명 2**
$X$와 $X_{1}$사이에 아래 관계가 성립한다    
> $X$는 $X_{1}$을 반복적이면서 독립적으로 $n$번 실행하는 확률 분포    
➡️ $X_{1}$을 $n$번 독립관측하는 것

**결론**    
$X=X_1+⋯+X_1$    $_{(X_1\ \ 을 \ \ n번 더하기)}$

### **설명 3**
위 식을 이용하여 $E(X)$와 $Var(X)$를 아래와 같이 계산할 수 있다.

**결론**    
- $E(X) = E(X_{1}+⋯+X_{1})$    
  $\ \ \ \ \ \ \ \ \ \ =n⋅E(X_1)$    
  $\ \ \ \ \ \ \ \ \ \ =n⋅p$
- $Var(X) = Var(X_1+⋯+X_n)$      
  $\ \ \ \ \ \ \ \ \ \ \ \ \ =n ⋅ Var(X_1)$     
  $\ \ \ \ \ \ \ \ \ \ \ \ \ =n⋅p⋅q$


✏️ **이항분포 예제**)    
1. 동전을 $n$번 던져 앞면이 $r$번 나올 확률들의 분포    
$X_{coin} ∼ B(n, \frac{1}{2})$
2. 주사위를 $n$번 던져 2보다 큰 소수가 $r$번 나올 확률들의 분포    
$X_{dice} ∼ B(n, \frac{1}{3})$
$_{이유\ :\ 2보다\ 큰\ 소수는\ 3과\ 5이다}$

## **푸아송 분포**
특정 기간 동안 어떤 사건이 발생하는 횟수의 평균값 또는 비율이 주어졌을 때, 미래의 동일한 기간동안 해당 사건이 $r$번 발생할 확률들의 분포


✏️ **예제**) 영화관의 낡은 팝콘 기계    
- 주중$_{(특정\ 기간)}\ \ \ \ $에 평균 3.4회$_{(사건\ 발생\ 빈도수의\ 평균값)}\ \ \ \ \ \ \ \ \ \ $ 고장 발생

```
질문❔   
다음 주중에 한 번도 고장이 발생하지 않을 확률은?
```

```
해법❕     
이런 경우 푸아송분포를 활용한다.    
팝콘 기계의 경우 아래처럼 표기한다.
```
$$X ∼ P_0(3.4)$$

---

###**일반화**
어떤 주어진 구간에 사건이 발생하는 수를 $X$라고 할때, 만약 $X$가 구간 마다 $λ_{(람다라\ 읽는다)}\ \ \ \ \ 번$ 발생하는 평균값을 갖는다면,     
$X$를 **푸아송 분포를 따른다**고 말하고 아래와 같이 표기한다.    
$$X ∼ P_0(λ)$$
$X$가 푸아송 분포를 따를 때 특정 구간에 언급된 사건이 $r$번 발생할 확률, 즉 $P(X=r)$은 아래와 같다.
$$P(X=r) = \frac{{e}^{-λ}⋅λ^{r}}{r!}$$
$$_{(e 	\simeq 2.71,\ 오일러\ 상수)}$$

---

✏️ **예제**) 영화관 팝콘 기계    
영화관 팝콘 기계의 고장율은 푸아송 분포$X ∼ P_0(3.4) $를 따른다.    
따라서, 예를 들어 다음 주에 팝콘 기계가 한 번도 고장나지 않을 확률은 다음과 같다.    
$$P(X=0) = \frac{e^{-3.4}⋅(3.4)^0}{0!}$$
$$\ \ \ = \frac{1}{e^{3.4}} ≈ 0.033$$
즉, **3%** 정도이다.

**결론** : 아무래도 새로운 팝콘 기계를 준비하는게 좋을 듯 싶다.

---

##**푸아송 분포의 기대치와 분산**
푸아송 분포의 기대치와 분산은 매우 간단하다.

$X ∼ P_0(λ)$인 경우    
- $E(X) = λ$
- $Var(x) = λ$
> $E(X) = λ$ ➡️ 상식적으로 생각할 수 있는 값과 일치    
∵ 경험적으로 특정 기간에 평균 $λ$번 어떤 사건이 발생하였다면,     
미래에도 동일한 기간동안 $λ$번 동일한 사건이 발생할 것이라 기대할 수 있다.

---

##**푸아송 분포의 결합**
$X$와 $Y$가 독립확률변수일 때 아래가 성립한다(*5장 참조*)
$$E(X+Y)=E(X)+E(Y)$$
따라서, 두 개의 독립 확률 변수 $X,\ Y$에 대해서도 이해관계가 성립한다.
$$X ∼ P_0(λ_{x})$$
$$Y ∼ P_0(λ_{y})$$
$$_{두\ 사건이\ 서로\ 독립이며\ 동일한\ 기간을\ 대상으로\ 해야함}$$

$$\Rightarrow \begin{pmatrix} 
X+Y ∼ P_0(λ_{x}+λ_{y})    \\
E(X+Y) = Var(X+Y) = λ_x + λ_y
\end{pmatrix} $$

--- 

✏️ **예제**)     
- 팝콘 기계 : $X ∼ P_0(3.4)\ \ \ _{(3.4\ :\ 팝콘\ 기계\ 일주일동안\ 고장\ 횟수)}$ 
- 음료수 기계 : $Y ∼ P_0(2.3)\ \ \ _{(2.3\ :\ 음료수\ 기계\ 일주일동안\ 고장\ 횟수)}$
> *서로 독립*   

두 기계중 어느 하나의 일주일 동안의 고장 횟수는 $X+Y ∼ P_0(3.4+2.3)$의 푸아송 분포를 따른다.    
따라서, 예를 들어 다음주 일주일 동안 어느 기계도 고장나지 않을 확률은 아래와 같다. 
$$P(X+Y=0) = \frac{e^{-5.7}⋅5.7^{0}}{0!} = 0.003 = 0.3\% $$

---


##**이항분포와 푸아송 분포의 관계**
$X∼B(n,p)$이고 $n \geq 50,\ p \leq 0.1$인 경우 $X ∼ P_0(np)$라고 해도 무방하다.    
즉, 이항분포인 경우 $n$이 적당히 크고, $p$가 꽤 작으면 푸아송 분포를 이용하여 확률을 계산할 수 있다.    

✏️ **예제**)    
```
질문 ❔      
50개의 문제를 무작정 찍어서 5개의 답을 맞출 확률은 얼마인가?     
단, 한 문제를 찍어 정답을 맞출 확률은 0.05이다.
````

```
견본 답안 ❕    
X가 정답 개수라 할때, X~B(50, 0.05)이다. 따라서 푸아송분포를 이용할 수 있다.
```
$$X ∼ P_0(2.5)$$

$$\Rightarrow P(X=5) = \frac{e^{-2.5}⋅(2.5)^{5}}{5!}=0.067=6.7\% $$
