### 3. 모집단의 평균 및 분산 추정하기
__① 점추정__
- 표본을 통해 구한 정보(평균, 분산 등)를 이용하여 모집단에 대한 정보를 구하는 방법  
  
  ▶ 점추정을 이용하여 구한 모집단에 대한 정보(평균, 분산 등)가 정확하다고 장담할 수는 없다. 하지만 그렇게 하는 것이 최선이다.  
 
 주의 : 편향되지 않은 표본이 구성되도록 최선을 다해야 한다.

$  $

__② 모집단 평균값의 점추정__  
ⅰ) 표본의 평균값 x̄를 구한다.  
- $x̄ = \frac{Σx}{n}$
  - $Σx$ : 표본에 속한 모든 데이터 더하기
  - $n$ : 표본의 크기

ⅱ) 모잡단의 평균값이 추정치를 설정한다.
- $μ̂=x̄$

즉, 표본의 평균을 모집단 평균의 추정치로 사용. 
  
  - 주의 : 모집단의 진짜 평균은 $μ$로 표시한다. 하지만 μ는 일반적으로 거의 알 수가 없으며 평균값의 추정치인 $μ̂$를 μ대신에 사용한다.

▶ 보다 정확한 μ̂를 얻기 위해 편향되지 않으면서 __적당한__크기의 표본을 구성하는 일이 매우 중요하다.  

$  $

__③ 모집단 분산의 점추정__  

평균값의 점추정과는 달리 모집단 분산의 점추정은 표본의 분산과 다르게 계산한다.  
$s^2 = \frac{Σ(x-x̄)^2}{n-1}$  
- $s^2$ : 모집단 분산의 점추정 값으로 사용하는 값
- $x̄$ : 표본의 평균값
- $n-1$ : 표본의 크기. 
  - $n$대신 $n-1$을 사용.  
  (이유 : 모잡단의 분산은 표본 분산에 비해 좀 더 큰 값을 가짐)

▶ 모집단 분산의 점추정 : $σ̂^2 = s^2$

$  $

__✏️예제__  
"달콤한 풍선껌" 회사의 제품의 표본에 속한 껌들의 맛의 지속시간이 다음과 같다. 


```
(단위 : 분)
61.9    62.6    63.3    64.8    65.1
66.4    67.1    67.2    68.7    69.9
```

▶ $μ̂ = x̄ = \frac{61.9+62.6+…+69.9}{10} = 65.7
$  
▶ $σ̂^2 = s^2 = \frac{(61.9-65.7)^2 + … + (69.9-65.7)^2}{9} = 6.92$

### 5. 평균의 표본분포 예제
__✏️ 예제__  
- 풍선껌을 봉지에 담아 판매
- 한 봉지에 담긴 풍선껌 개수의 평균은 10, 분산은 1임.
- 한 상자에 30봉지씩 넣어 판매함.


__질문__ :  
__①__ 한 상자에 들어있는 풍선껌 한 봉지에 담긴 풍선껌 개수의 평균 값은 얼마인가 ?   
__②__ 한 상자에 들어있는 풍선껌 한 봉지에 담긴 풍선껌 개수의 평균이 8.5개 이하일 확률은 ?

$  $

예제의 질문에 답하기 위해서는 아래 두 과정을 진행해야 한다.  
__①__ 30개의 봉지를 무작위로 선택하여 하나의 봉지에 들어있는 풍선껌 개수의 평균을 구해보아야 한다.  
__②__ 과정①을 여러번(일반적으로 30회 이상) 반복하여 풍선껌 개수 평균의 분포를 알아낸다.  

__주의사항__ : 30개의 선택은 매번 독립적으로 이루어져야 한다. 즉, 한 봉지의 선택은 무작위적으로 이루어지며 한 번 선택된 봉지를 다시 원래대로 돌려 넣는 __복원추출__ 방식을 따른다. 

###### __예제 답안__


$X$를 한 봉지에 들어있는 풍선껌 개수의 분포를 나타낸다고 하자. 그러면 다음 사실이 성립한다.  
__①__ $E(X) = 10$  
__②__ $Var(X) = 1$  
__③__ 봉지 30개 각각의 선택은 __독립관측__이다.

$  $

따라서 30봉지에 들어있는 풍선껌 개수의 평균값의 분포를 $X̅$라 하면 아래 공식이 성립한다.  
__①__ $X̅ = \frac{X_1 + ⋯ + X_30}{30}$  
__②__ $X_i$는 $X$의 독립관측

$  $

$X_i$가 $X$의 독립관측이므로 X̅의 평균값과 분산을 다음과 같이 구할 수 있다.  
- $E(X) = E(\frac{X_1 + … + X_{30}}{30})$   
$= E(\frac{1}{30}X_1 + … + \frac{1}{30}X_{30})$  
$= \frac{1}{30}E(X_1) + ⋯ + \frac{1}{30}E(X_{30})$  
$= \frac{1}{30}(E(X_1)+⋯+E(X_{30}))$  
$= \frac{1}{30} ⋅ 30 ⋅ E(X)$  
$= E(X) = 10$  

- $Var(X̅) = Var(\frac{X_1 + … + X_{30}}{30})$  
$= Var(\frac{1}{30}X_1 + … + \frac{1}{30}X_{30})$  
$= \frac{1}{30^2}Var(X_1) + … + \frac{1}{30^2}Var(X_{30})$  
$=  \frac{1}{30^2} ⋅ 30 ⋅ Var(X)$  
$= \frac{Var(X)}{30} = \frac{1}{30}$

위 결과는 상식적으로 이해 가능하다.  
분산의 경우: 한 봉지에 포함된 풍선껌의 개수가 늘어날 수록 분포가 비례하여 상쇠됨.


이제 $P(X̅<8.5)$를 구하고자 한다.  
그런데 X̅는 어떤 분포를 따르는가 ?
- 정규분포를 따른다.  
즉, $X̅ \sim N(10, \frac{1}{30})$이 성립한다.  
이유 : __중심 극한 정리__ (뒤에서 설명된다.) 

따라서 $P(X̅<8.5) = P(z<\frac{8.5-10}{\sqrt{0.0333}}) = P(z<-8.22) = 0$
- 확률테이블에서 확인이 불가하므로 사실상 0임을 의미한다.
- __주의__ : 연속성보정 필요없음
  - 평균값들은 어떠한 값도 임의도 가질 수 없기 때문.

### 6. 평균의 표본분포(일반화) 

$E(x) = \mu,\ Var(X)=\sigma^2$을 따르는 $X$의 분포가 주어짐. $X$를 $n$번 독립관측 했을 때, 관측된 $X$들의 평균 값을 가리키는 확률변수를 X̅로 표기한다. 그러면 아래 식이 성립한다.  
  X̅$=\frac{X_1+…X_n}{n}$  
  ($X_i$는 $X$의 독립관측)

또한 X̅의 평균값과 분산은 다음과 같다.  
- $E($X̅$) = E(\frac{X_1+…X_n}{n})$  
$= \frac{1}{n}\cdot E(X_1+…X_n)$  
$=\frac{1}{n}\cdot n \cdot E(X)$  
$=E(X) = μ$
- $Var($X̅$) = Var(\frac{X_1+…X_n}{n})$  
$= \frac{1}{n^2}\cdot E(X_1+…X_n)$  
$=\frac{1}{n^2}\cdot n \cdot Var(X)$  
$=\frac{1}{n}Var(X) = \frac{σ}{n}$
  - 일반적으로 $n$이 커질 수록 표준오차는 줄어든다.

특히 $n≥30$ 일 때 X̅는 <U>중심극한정리에 의해</U> 정규분포를 따른다. 즉,  
$X̅∼N(μ, \frac{σ^2}{n})$  
(단, $E(X)=μ, Var(X) = σ^2)$  
- __주의__ : $X$가 반드시 정규분포를 따를 필요는 없다.


### 7. 중심극한정리

__①__ 서로 독립이며 동일한 분포를 따르는 $n$개의 확률변수 $X_1, X_2,…,X_n$에 대해 $n$이 충분히 크면, 예를 들어 $n≥30$, 아래와 같이 정의된  
X̅$=\frac{X_1+…X_n}{n}$  
X̅의 분포가 개략적으로 정규분포를 따른다.  
또한, $E(X)=μ$이고 $Var(X) = σ^2$이고 $X_i$가 $X$와 동일한 확률분포를 가지면 다음이 성립한다.  
X̅$∼N(μ, \frac{σ^2}{n})$   
(단, $n≥30$)  

$  $

__②__ 확률찾기  
X̅가 정규분포를 따르므로 확률테이블을 이용하여 확률값을 계산할 수 있다.  
__(ⅰ)__ $P($X̅$>a=P(z>\frac{a-\mu}{\frac{σ}{\sqrt{n}}})$  
__(ⅱ)__ $P(a<$X̅$<b)=P(\frac{a-\mu}{\frac{σ}{\sqrt{n}}}<z<\frac{b-\mu}{\frac{σ}{\sqrt{n}}})$  
__(ⅲ)__ $P(a<$X̅)$=P(\frac{a-\mu}{\frac{σ}{\sqrt{n}}}<z)$  
(__주의__ : 연속성 보정 필요 없음)  
- 평균값들은 어떠한 값도 임의로 가질 수 있기 때문. 

### 8. 중심극한정리 활용  
<U>__1)이항분포__</U>  
전제 : $X ∼ B(m, p)$  
▶ $\mu = m ⋅ p$, $σ^2 = n⋅p⋅q$ $(q = 1-p)$  
이항분포를 따르는 $X$를 독립관측으로 $n$번 $(n≥30)$ 반복할 때 중심극한정리에 의해 $X$의 평균값 X̅는 정규분포를 따른다. 즉,  
X̅$\sim N(m \cdot p, \frac{n⋅p⋅q}{n})$  

$  $

__✏️예제)__ 동전을 10번 던졌을 때 숫자가 나오는 횟수는 이항분포를 따른다. 즉,  $X ∼ B(10, \frac{1}{2})$.  
- __주의__ : $X$자체는 정규분포를 따르지 않는다.

이제, "동전 10번 던지기"를 30번 반복했을 때 숫자가 나오는 횟수의 평균은 다음의 분포를 따른다.  
$X ∼ B(10\cdot\frac{1}{2} = 5, \frac{10\cdot\frac{1}{2}\cdot\frac{1}{2}}{30} = \frac{1}{12})$  

$  $

__✏️예제)__ "동전 10번 던지기"를 30번 반복했을 때, 숫자가 평균 6번 이상 나올 확률은 ?  
- 정답 : X̅ $∼N(5, \frac{1}{12})$이므로  
$P($X̅$>6 = P(z>\frac{6-5}{\frac{1}{\sqrt{12}}})$  
$= P(z>\sqrt{12}) = P(z>3.46)$  
$= 1-P(z ≤ 3.46)$  
$ = 1-0.999730$  
$ ≃ 0.000270$ (매우 낮음)

  ▶ $X$가 동전을 300번 던져서 숫자가 나오는 횟수를 가리킨다면  
  이다.  
  그리고 $300 × \frac{1}{2} = 150>5$이므로 $X$는 정규 분포를 따른다. 즉,  
$X \sim N(150, 75)$  
따라서  
$P(X>180) = P(z>\frac{180.5-150}{\sqrt{75}})$  
$= P(z>3.52)$  
$= 1-P(z<3.52)$  
$= 1-0.999784$  
$≃ 0.000216$
    - 0.00027과 비교하여 오차는 0.000054. 즉, 0.0054%임. 확률로는 어떤 의미도 없는 오차임.

<U>__2) 푸아송 분포__</U>(다루지 않음)  
전제 : $X \sim P_0(λ)$  
▶ $μ=λ$, $σ^2 = λ$  
푸아송분포를 따르는 $X$를 독립관측으로 $n$번$(n \ge 30)$반복할 때 중심극한정리에 의해 $X$의 평균값 X̅는 정규분포를 따른다. 즉,  
X̅ $\sim N(λ, \frac{λ}{n})$  

$  $

__✏️예제__ 7장에서 살펴본 영화관의 팝곤기계의 주당 평균 고장 횟수는 푸아송분포를 따르며 $X \sim P_0(3.4)$이다.    
- __주의__ : $X$자체는 정규분포를 따르지 않는다.  

이제, 30주 동안 주당 고장 횟수의 평균은 다음의 정규분포를 따른다.:  
X̅ $\sim N(3.4,0.11 = \frac{3.4}{30})$

$  $

__✏️예제__ 임의로 지정된 30부동안 주당 고장횟수의 평균이 2회 이하일 확률은 ?  
- __정답__ : X̅ $\sim N(3.4,0.11)$ 이므로  
$P($X̅$<2) = P(z<\frac{2-3.4}{\sqrt{0.113}})$  
$= P(z<\frac{-1.4}{0.337})$  
$= P(z<-4.15)$  
$= 0$ ➡ 거의 0임
